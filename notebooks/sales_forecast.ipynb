```python
#1: Imports
import pandas as pd
import numpy as np
import os
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error


#2: Functions

def load_sales_data(filepath='data/processed_walmart_sales.csv'):
    df = pd.read_csv(filepath, parse_dates=['Date'])
    return df

def get_unique_store_dept_pairs(df):
    return df[['Store', 'Dept']].drop_duplicates().reset_index(drop=True)

def filter_store_dept_data(df, store, dept):
    subset = df[(df['Store'] == store) & (df['Dept'] == dept)].copy()
    subset = subset.sort_values('Date')
    return subset[['Date', 'Weekly_Sales']].reset_index(drop=True)

def evaluate_forecast(y_true, y_pred):
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return round(mape, 2), round(rmse, 2)

def forecast_with_prophet(df_subset, store, dept):
    df_prophet = df_subset.rename(columns={'Date': 'ds', 'Weekly_Sales': 'y'})
    split_index = int(len(df_prophet) * 0.8)
    train_df = df_prophet.iloc[:split_index]
    test_df = df_prophet.iloc[split_index:]

    model = Prophet(yearly_seasonality=True, weekly_seasonality=True)
    model.fit(train_df)

    future = df_prophet[['ds']]
    forecast = model.predict(future)

    y_pred = forecast.iloc[-len(test_df):]['yhat'].values
    y_true = test_df['y'].values
    mape, rmse = evaluate_forecast(y_true, y_pred)

    result = df_prophet.copy()
    result['yhat'] = forecast['yhat']
    result['Store'] = store
    result['Dept'] = dept

    return result, mape, rmse


#3: Loop Across All Store–Dept Pairs

def run_forecasting_pipeline(data_path='data/processed_walmart_sales.csv', output_dir='forecasts'):
    os.makedirs(output_dir, exist_ok=True)

    df = load_sales_data(data_path)
    store_dept_pairs = get_unique_store_dept_pairs(df)

    results = []

    for _, row in store_dept_pairs.iterrows():
        store, dept = row['Store'], row['Dept']
        subset = filter_store_dept_data(df, store, dept)

        if len(subset) < 30:
            continue  # Skip short series

        try:
            forecast_df, mape, rmse = forecast_with_prophet(subset, store, dept)
            csv_path = os.path.join(output_dir, f'Store{store}_Dept{dept}_forecast.csv')
            forecast_df.to_csv(csv_path, index=False)

            results.append({
                'Store': store,
                'Dept': dept,
                'MAPE': mape,
                'RMSE': rmse,
                'Forecast CSV': csv_path
            })

            print(f"✅ Store {store} Dept {dept} | MAPE: {mape}% | RMSE: {rmse}")

        except Exception as e:
            print(f"❌ Store {store} Dept {dept} failed: {e}")

    summary_df = pd.DataFrame(results)
    summary_df.to_csv(os.path.join(output_dir, 'forecast_summary.csv'), index=False)
    print("📊 Forecasting complete. Summary saved.")


#4: Run pipeline
run_forecasting_pipeline()
```
